This notebook demonstrates Random Forest Classifier.

```
from sklearn.ensemble import RandomForestClassifier
import numpy as np

X = np.array([[1],[2],[3],[4],[5]])
y = np.array([0,0,1,1,1])

model = RandomForestClassifier()
model.fit(X,y)
print(model.predict(X))
```
