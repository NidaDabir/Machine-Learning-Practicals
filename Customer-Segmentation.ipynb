This notebook demonstrates Customer Segmentation using KMeans.

```
import pandas as pd
from sklearn.cluster import KMeans

# Example dataset
X = [[19,15,39],[21,15,81],[20,16,6],[23,16,77]]
model = KMeans(n_clusters=2)
model.fit(X)
print(model.labels_)
```
