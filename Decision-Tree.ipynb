This notebook demonstrates Decision Tree Classifier.

```
from sklearn.tree import DecisionTreeClassifier
import numpy as np

X = np.array([[1],[2],[3],[4],[5]])
y = np.array([0,0,1,1,1])

model = DecisionTreeClassifier()
model.fit(X,y)
print(model.predict(X))
```
